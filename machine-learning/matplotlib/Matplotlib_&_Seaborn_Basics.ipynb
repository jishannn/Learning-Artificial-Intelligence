```python
import matplotlib.pyplot as plt
# Seaborn is built on top of Matplotlib.
#Seaborn simplifies the process with built-in themes and better default styles.
import seaborn as sns
import numpy as np
import pandas as pd
```

**Basic Line Plot**

```python
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y = np.sin(x)

plt.plot(x, y, label='Sine Wave', color='b', linestyle='--', linewidth=2)
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Explanation
# .plot(x, y) → Creates a line plot.
# label="Sine Wave" → Adds a label.
# color='b' → Blue color.
# linestyle="--" → Dashed line.
# linewidth=2 → Line thickness.
# plt.legend() → Shows the legend.
# plt.grid(True) → Adds a grid.
```

**Bar Chart with Real-World Data**

```python
netflix_dataset = pd.read_csv('netflix_titles.csv')   # load the dataset
netflix_dataset.head()   # by default it returns first 5 rows of the dataframe or dataset or series.
```

Plot a Bar Chart

```python
country_counts = netflix_dataset['country'].value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.bar(country_counts.index, country_counts.values, color='red')

plt.title('Top 10 countries with Most Netflix Shows')
plt.xlabel('Country')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)
plt.show()
```

**Scatter Plot - Stock Market Data**

```python
apple_stock_dataset = pd.read_csv('AAPL.csv')
apple_stock_dataset.head()
```

Scatter Plot of Closing Prices

```python
plt.figure(figsize=(10, 5))
plt.scatter(apple_stock_dataset['Date'], apple_stock_dataset['Close'], color='green', alpha=0.5)   # alpha=0.5 makes the points semi-transparent.

plt.title('Apple Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

**Histogram - Covid-19 Cases**

```python
# load the dataset
covid_dataset = pd.read_csv('WHO-COVID-19-global-data.csv')
covid_dataset.head()
```

Histogram of Daily Cases

```python
plt.figure(figsize=(10, 5))
plt.hist(covid_dataset['New_cases'], bins=30, color='purple', edgecolor='black')
# bins=30 it indicates that the histogram will display the data distribution using 30 distinct groupings.
# In short, bins=30 defines 30 intervals.

plt.title('Distribution of Daily COVID-19 Cases')
plt.xlabel('Number of Cases')
plt.ylabel('Frequency')
plt.show()
```

**Pie Chart - Device Usage**

```python
labels = ['Mobile', 'Laptop', 'Tablet', 'Desktop']
sizes = [50, 30, 10, 10]   # Percentage share
colors = ['gold', 'lightblue', 'lightgreen', 'red']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
# autopct='%1.1f%%' shows percentage labels.

plt.title('Device Usage Distribution')
plt.show()
```

**Combining Seaborn & Matplotlib**

Bar Plot with Seaborn

```python
plt.figure(figsize=(10, 5))

sns.barplot(x=country_counts.index,
            y=country_counts.values,
            hue=country_counts.index,  # Assign x variable to hue
            palette="coolwarm",
            dodge=False,  # Prevents bars from being grouped
            legend=False)  # Disable legend as it's redundant

plt.title("Top 10 Countries with Most Netflix Shows", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Shows", fontsize=12)
plt.xticks(rotation=45, ha="right")  # Rotate labels for readability

plt.show()
```

**Subplots - Multiple Charts in One Figure**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Line Plot
axes[0].plot(x, y, color='blue')
axes[0].set_title('Sine Wave')

# Scatter plot
axes[1].scatter(apple_stock_dataset['Date'], apple_stock_dataset['Close'], color='red', alpha=0.5)
axes[1].set_title('Apple Stock Closing Prices')

plt.tight_layout()
plt.show()
```
